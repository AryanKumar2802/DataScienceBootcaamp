### Univariate Analysis

#### Price

```python
df.price
```

```python
sns.histplot(data=df, x='price')
```

Handling Outliers

```python
data = df[df['price'] < 1500]
sns.boxplot(data=data, x='price')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=data, x='price', bins=100)
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=df, x='availability_365')
plt.ylabel('Frequency')
plt.title('availability_365 Distribution')
plt.show()
```

```python
df.groupby(by="neighbourhood_group")['price'].mean()
```

### Feature Engineering

```python
# Creating a new column named price per bed
df['price per bed'] = df['price']/df['beds']
df.head()
```

```python
# df.shape
df['price per bed']
```

```python
df.groupby(by="neighbourhood_group")['price per bed'].mean()
```

price dependency on neighbourhood_group

number of reviews and price relationship

```python
plt.title("Locality and Review Dependncy")
sns.scatterplot(data=data, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=data, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue="room_type")
```

geographical distribution of airbnb listings

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of Airbnb Listings")
plt.show()
```

correlation of one variable with others for numerical column

```python
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
plt.figure(figsize = (8,6))
sns.heatmap(data=corr, annot=True)
```
